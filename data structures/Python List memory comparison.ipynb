## Python List Length - Dynamic memory Graph

### Python adds memory to the list dynamically by chunks when it detects overflow

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy
```

```python
import sys
```

```python
l = []
size = 1000
lenth = []
memory = []

for counter in range(size):
    l.append(counter)
    lenth.append(len(l))
    memory.append(sys.getsizeof(l))

# print(lenth)
# print(memory)

plt.plot(lenth, memory)
plt.title("Python List lenth - memory comparison")
plt.xlabel("Length")
plt.ylabel("Memory")
plt.show()
```
