```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
type(iris)
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx, testx = train_test_split(iris.data, test_size=0.1, random_state=1)
trainy, testy = train_test_split(iris.target, test_size=0.1, random_state=1)
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(trainx, trainy)
```

```python
predy = clf.predict(testx)
```

```python
predy == testy
```

## Visualization Code

```python
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names)
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```
